# Data
***

Scikit-Learn's **Wine Dataset**.

## Features

1. **Alcohol**: The alcohol content in the wine.
2. **Malic acid**: Concentration of malic acid.
3. **Ash**: Content of ash.
4. **Alcalinity of ash**: Measure of ash alkalinity.
5. **Magnesium**: Magnesium content.
6. **Total phenols**: Total phenolic content.
7. **Flavanoids**: Amount of flavonoids.
8. **Nonflavanoid phenols**: Amount of nonflavonoid phenolic compounds.
9. **Proanthocyanins**: Proanthocyanin content.
10. **Color intensity**: Intensity of the color of the wine.
11. **Hue**: Hue of the wine.
12. **OD280/OD315 of diluted wines**: Ratio of absorbance of diluted wines at 280 nm and 315 nm.
13. **Proline**: Concentration of proline amino acid.

## Target (Class Label)

**Type of wine cultivar**: Varieties the sample belongs to. 
* **Class 0**: First type of wine cultivar
* **Class 1**: Second type of wine cultivar
* **Class 2**: Third type of wine cultivar

# Importing the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
```

# Loading the data

```python
data = load_wine()
X = data.data
y = data.target
df = pd.DataFrame(X, columns=data.feature_names)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

*No null values.*

# 2 Components

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

## Explained Variance

```python
explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio:", explained_variance)
print("Total explained variance:", explained_variance.sum())
```

## Plot

```python
plt.figure(figsize=(20, 5))
custom_cmap = ListedColormap(['#fff44f', '#32de84', '#FF033E'])
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap=custom_cmap, edgecolor='k', s=100)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D PCA of the Wine Dataset')
plt.show()
```

# 3 Components

```python
pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X)
```

## Explained Variance

```python
explained_variance = pca_3d.explained_variance_ratio_
print("Explained variance ratio:", explained_variance)
print("Total explained variance:", explained_variance.sum())
```

## Plot

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
custom_cmap = ListedColormap(['#fff44f', '#32de84', '#FF033E'])
sc = ax.scatter(X_pca_3d[:, 0], X_pca_3d[:, 1], X_pca_3d[:, 2], c=y, cmap=custom_cmap, s=100, edgecolor='k')
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')
ax.set_title('3D PCA of the Wine Dataset')
plt.show()
```
